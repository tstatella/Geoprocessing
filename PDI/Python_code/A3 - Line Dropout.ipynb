```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np # as imagens são lidas como numpy arrays
from matplotlib import pyplot as plt # é preciso chamar os submódulos
from skimage import io
```

```python
img = io.imread('menina_dropout.tif')
print(f"Classe: {type(img)}")
print(f"Tipo de variável: {img.dtype}")
print(f"Min, max: {img.min(),img.max()}")
print(f"Dimensões: {img.shape}")
plt.imshow(img,cmap='gray')
plt.colorbar(); 
```

```python
ind = np.where(img==0)
print(ind)

img_copy = img.copy()

for lin in np.arange(0,img.shape[0]):
    for col in np.arange(0,img.shape[1]):
        if img[lin,col] == 0:
            img_copy[lin,col] = img[lin-1,col]

plt.imshow(img_copy,cmap='gray'); 
plt.colorbar(); 
```
